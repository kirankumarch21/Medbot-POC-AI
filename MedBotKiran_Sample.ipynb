```python
# Add Azure OpenAI package
from openai import AzureOpenAI
import requests
from IPython.display import Image

def createTextBasedonText(azure_oai_endpoint, azure_oai_key, userPrompt, history):
  # print('startted createTextBasedonText')
  client = AzureOpenAI(
        azure_endpoint = azure_oai_endpoint,
        api_key=azure_oai_key,
        api_version="2024-02-15-preview"
        )
  # print('Created client for provided keys')

  history.append({"role": "user", "content": userPrompt})
  # print('History prepared for the conversation')
  # print('sending a request to openai for response')

  response = client.chat.completions.create(
    model=azure_oai_deployment,
    temperature=0.7,
    max_tokens=1200,
    messages=history
  )
  # print('Response received..')

  generated_text = response.choices[0].message.content

  # Add generated text to messages array
  history.append({"role": "system", "content": generated_text})
  return generated_text


def createImageBasedonPrompt(azure_endpoint, api_version, api_key, userPrompt):
  url = "{}openai/deployments/dalle3/images/generations?api-version={}".format(azure_endpoint, api_version)
  headers= { "api-key": api_key, "Content-Type": "application/json" }
  body = {
              "prompt": userPrompt,
              "n": 1,
              "size": "1024x1024"
          }
  response = requests.post(url, headers=headers, json=body)
  print(response)
  revised_prompt = response.json()['data'][0]['revised_prompt']
  image_url = response.json()['data'][0]['url']

  # Display the URL for the generated image
  print(revised_prompt)
  print(image_url)
  save_path = 'generated_image.jpg'

  # Download the image and save it to the specified path
  response = requests.get(image_url)
  if response.status_code == 200:
      with open(save_path, 'wb') as f:
          f.write(response.content)
      print(f"Image saved to {save_path}")
      Image(save_path)
  else:
      print("Failed to download the image")


prompt = input("\nEnter a prompt: ")

isImageOrText = input("\nOutput in format \n 1.Text \n 2.Text and Image \n\n")
print("\n Selected Option is " + isImageOrText + "\n\n")
while True:
  if isImageOrText == '1':
    print('Generating the text for the provided prompt...')

    azure_oai_endpoint ="https://eyuser23.openai.azure.com/"
    azure_oai_key ="e658a917fc2c47cf9042d085c8a295b9"
    azure_oai_deployment ="Kiran"
    history = [ {"role": "system", "content": "You are a helpful assistant with Medical Bot with medical background."},
        {"role": "user", "content": "Provided suggestion on medicine or a problem should be self explanatory for resolving problem specified"},
        {"role": "assistant", "content": "Yes, Provided suggestions are checked before sending back"},
        {"role": "user", "content": "Is it possible to send detail description with problem and solution"},
        {"role": "user", "content": "All the prompts from users should be related medical or medicines only, if not please let user know to prepare the prompt in respective way"} ,
        {"role": "user", "content": "Provide a solution like a doctor"},
        ]
    generated_text = createTextBasedonText(azure_oai_endpoint, azure_oai_key, prompt, history)
    print('Med Assistance: \n')
    print(generated_text)
    break

  elif isImageOrText == '2':
    azure_oai_endpoint_image ="https://eyuser26.openai.azure.com/"
    azure_oai_key_image ="98138d9733f746a6aa3652f29c183d0d"
    azure_oai_deployment_image ="Dalle3"
    api_version="2024-02-15-preview"
    context = "You are a helpful assistant with Medical Bot with medical background. Provide a solution like a doctor. Is it possible to send detail description with problem and solution. Always generate image with specifics only nothing else in image"


    azure_oai_endpoint ="https://eyuser23.openai.azure.com/"
    azure_oai_key ="e658a917fc2c47cf9042d085c8a295b9"
    azure_oai_deployment ="Kiran"
    history = [ {"role": "system", "content": "You are a helpful assistant with Medical Bot with medical background."},
        {"role": "user", "content": "Provided suggestion on medicine or a problem should be self explanatory for resolving problem specified"},
        {"role": "assistant", "content": "Yes, Provided suggestions are checked before sending back"},
        {"role": "user", "content": "Is it possible to send detail description with problem and solution"},
        {"role": "user", "content": "All the prompts from users should be related medical or medicines only, if not please let user know to prepare the prompt in respective way"} ,
        {"role": "user", "content": "Provide a solution like a doctor"},
        ]
    generated_text = createTextBasedonText(azure_oai_endpoint, azure_oai_key, prompt, history)
    print('Med Assistance: \n')
    print(generated_text)

    updatedPrompt = generated_text + "Return yes or no, accordingly for a medical question yes and for all other questions not relvant to medicine or medical return no"
    isYesorNo = createTextBasedonText(azure_oai_endpoint, azure_oai_key, updatedPrompt, history)
    print(isYesorNo)
    if isYesorNo == 'yes' or isYesorNo == 'Yes':
      createImageBasedonPrompt(azure_oai_endpoint_image, api_version,  azure_oai_key_image, prompt )
    else:
      print('Image can not be generated the prompt, please try again with different prompt')
    break

  else:
    print('Invalid input, please try again')
    break

```
